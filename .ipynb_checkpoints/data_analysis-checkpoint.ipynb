## Exploring & Analysing Data

```python
company.sample(15)
```

```python
x = company.groupby(['New Expansion'])
```

```python
company[company['New Expansion'] == 'Old'].groupby(['State']).sum()['Revenue'].nlargest(10)
```

```python
company[company['New Expansion'] == 'New'].groupby(['State']).sum()['Revenue'].nlargest(10)
```

```python
company['Profit'] = company['Revenue'] - company['Marketing Spend']
```

```python
company
```

```python
company['ROMS'] = round((company['Profit'] / company['Marketing Spend']) * 100, 2)
```

```python
company['ROMS%'] = round((company['Profit'] / company['Marketing Spend']) * 100, 2)/100
```

```python
company
```

```python
company.to_csv('startups-expansion-modified.csv')
```

## Visualisation

```python
# Sales Region 
company['Sales Region'].value_counts().plot.bar()
```

```python
# profil by state and sales region 

Profit = company["Profit"]
Store = company["Store ID"]
marketing = company["Marketing Spend"]
plt.bar(Store,Profit)
plt.bar(Store,marketing, bottom=Profit)

plt.xlabel("Profit and marketing spend")
plt.ylabel("Store ID")
plt.title("Profit and marketing spend by Store ID")
```
